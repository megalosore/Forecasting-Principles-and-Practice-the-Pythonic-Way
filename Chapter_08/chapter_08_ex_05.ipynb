```python
df = pd.read_csv("../DATA/global_economy.csv")
df = df[df.unique_id == "France"]
exports = df[["unique_id", "ds"]]
exports["y"] = df["Exports"]
```

```python
AutoETS(season_length=1, model="ANN").fit(exports.y.values)

plot_series(exports)
```

```python
sf = StatsForecast(models=[AutoETS(model="ANN", season_length=1)], freq=1)

fc = sf.forecast(df=exports, h=10, level=[80, 95], fitted=True)
plot_series(
    exports,
    fc,
    level=[80, 95],
    xlabel="Year",
    ylabel="Exports",
    title="France Yearly Exports",
    rm_legend=False,
)
```

```python
insample = sf.forecast_fitted_values()
root_mean_squared_error(insample["y"], insample["AutoETS"])
```

```python
resid = insample["y"] - insample["AutoETS"]
sns.histplot(resid)
```

```python
plot_acf(resid)
```
