```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(f'../DATA/tute1.csv')
df['Quarter'] = pd.to_datetime(df['Quarter'])
df.head(5)
```

```python
fig, axes = plt.subplots(3,1)
cols = ['Sales', 'AdBudget', 'GDP']
for col, ax in zip(cols, axes.flatten()):
    sns.lineplot(df, x='Quarter', y=col, ax=ax)
```

```python
df = pd.read_csv(f'../DATA/us_total.csv')
df['ds'] = pd.to_datetime(df['ds'], format='%Y')

fig, axes = plt.subplots(3,2, figsize=(12,12))
states = ['Maine', 'Vermont', 'New Hampshire', 'Massachusetts', 'Connecticut', 'Rhode Island']
for state, ax in zip(states, axes.flatten()):
    sns.lineplot(df[df['unique_id'] == state], x='ds', y='y', ax=ax)
    ax.set_ylabel(f'Gaz Demand in {state}')
plt.tight_layout()
```
