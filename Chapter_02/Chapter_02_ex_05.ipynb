```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(f'../DATA/tourism.csv')
df['ds'] = pd.to_datetime(df['ds'])
groups = df.groupby(['Region', 'Purpose'])
```

```python
groups['y'].mean().idxmax()
```

```python
state_df = df.groupby(['State', 'ds'])['y'].sum().reset_index()
```

```python
fig, axes = plt.subplots(3,3, figsize=(12,12))
for state, ax in zip(state_df.State.unique(), axes.flatten()):
    sns.lineplot(state_df[state_df['State'] == state], x='ds', y='y', ax=ax)
    ax.set_ylabel(f'Tourism in {state}')
plt.tight_layout()
```
