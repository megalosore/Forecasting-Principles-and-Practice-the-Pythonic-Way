```python
df = pd.read_csv(f'../DATA/PBS.csv')
df['Month'] = pd.to_datetime(df['Month'])
df = df.drop('Unnamed: 0', axis=1)
df = df.set_index('Month')
df = df[df['Concession'] == 'General']
df.head()
```

```python
df['ATC_MEAN'] = df.groupby('ATC2')['Cost'].transform('mean')
df['ATC_STD'] = df.groupby('ATC2')['Cost'].transform('std')
```
